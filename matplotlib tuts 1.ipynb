```python
Population=[2.0,3.4,4.1,5.8,6.2]
year=[2000,2001,2002,2003,2004]
plt.plot(year,Population)
plt.show()
```

```python
y=[2000,2001,2002,2003]
p=[2.1,3.9,4.3,4.9]
plt.plot(y,p)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('world population')
plt.yticks([1,2,3,4,5],['1b','2b','3b','4b','6b'])
plt.legend("Ground")
plt.fill_between(y,p,0,color='green')


plt.show()
```

```python
d=200
plt.scatter(y,p,s=d*6)

plt.show()
```
